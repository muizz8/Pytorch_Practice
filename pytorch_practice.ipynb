````python
def solution(nums, target):
    n = len(nums)
    if n == 1:
        return 1


    dp = [1] * n

    # Iterate over each pair (i, j) where j < i
    for i in range(1, n):
        for j in range(i):
            if (nums[i] + nums[j]) % target != 0:
                dp[i] = max(dp[i], dp[j] + 1)


    return max(dp)


print(solution([1, 7, 2, 4], 3))
print(solution([1], 3))
````

##Derivatives

````python
x = torch.tensor(2.0, requires_grad=True)
y = x**2
y.backward()
print(x.grad)
````

In this code, the process of automatic differentiation in PyTorch is demonstrated. Here's a step-by-step explanation:

1. **Tensor Creation with Gradient Tracking:**
   ```python
   x = torch.tensor(2.0, requires_grad=True)
   ```
   A tensor `x` is created with a value of `2.0`. The `requires_grad=True` flag indicates that PyTorch should keep track of all operations on this tensor, allowing for automatic differentiation.

2. **Operation on Tensor:**
   ```python
   y = x**2
   ```
   A new tensor `y` is created by squaring `x`. This operation is tracked by PyTorch, enabling it to compute gradients later.

3. **Backward Propagation:**
   ```python
   y.backward()
   ```
   This function call computes the gradient of `y` with respect to `x`. Since `y = x**2`, the derivative of `y` with respect to `x` is `2x`. At `x = 2.0`, this derivative evaluates to `4.0`.

4. **Printing the Gradient:**
   ```python
   print(x.grad)
   ```
   The gradient of `y` with respect to `x`, which is stored in `x.grad`, is printed. The expected output is `4.0`.

Here's a summary of the mathematical operations:
- \( y = x^2 \)
- \(\frac{dy}{dx} = 2x \)
- Evaluating at \( x = 2.0 \), \(\frac{dy}{dx} = 4.0 \)

So, when `y.backward()` is called, PyTorch computes this gradient and stores it in `x.grad`, resulting in the output `4.0`.

````python
x = torch.tensor(2.0,requires_grad=True)
z = x**2 + 2*x + 1
z.backward()
print(x.grad)
````

The image shows the differentiation of a function \( z(x) \). Let's break it down to understand why we add 2 to \( 2x \) during differentiation.

1. **Function Definition:**
   \[
   z(x) = x^2 + 2x + 1
   \]

2. **Evaluation of the Function at \( x = 2 \):**
   \[
   z(2) = 2^2 + 2(2) + 1 = 4 + 4 + 1 = 9
   \]

3. **Derivative of the Function:**
   To find the derivative of \( z(x) \) with respect to \( x \):
   \[
   \frac{dz}{dx} = \frac{d}{dx}(x^2 + 2x + 1)
   \]
   We apply the power rule for differentiation:
   - The derivative of \( x^2 \) is \( 2x \).
   - The derivative of \( 2x \) is \( 2 \).
   - The derivative of a constant (1) is 0.

   Combining these, we get:
   \[
   \frac{dz}{dx} = 2x + 2
   \]

4. **Evaluating the Derivative at \( x = 2 \):**
   \[
   \frac{dz}{dx} \bigg|_{x=2} = 2(2) + 2 = 4 + 2 = 6
   \]

In summary, when differentiating \( z(x) = x^2 + 2x + 1 \), the term \( 2x \) comes from the derivative of \( x^2 \), and the term \( 2 \) comes from the derivative of \( 2x \). Thus, we add \( 2 \) to \( 2x \) in the derivative formula.

````python
u = torch.tensor(1.0,requires_grad=True)
v = torch.tensor(2.0,requires_grad=True)
z = u*v + u**2
z.backward()
print(u.grad)
print(v.grad)
````

# Build Custom Modules

````python
from torch import nn
from torch.nn import Linear
class Lr(nn.Module):
  def __init__(self,input_size,output_size):

    #inherit from parent
    super(Lr,self).__init__()
    self.linear = nn.Linear(input_size,output_size)

  def forward(self,x):
    yhat = self.linear(x)
    return yhat
````

````python
lr = Lr(1,1)
print("the parameter",list(lr.parameters()))
print("linear model",lr.linear)
````

````python
import torch
x = torch.tensor([[1.0]])
yhat = lr(x)
print("the prediction is",yhat)
````
